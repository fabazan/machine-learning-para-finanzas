```python
accion_hist = yf.download("SPY", "2018-01-01", "2022-04-29", interval="1d")
```

```python
accion_hist.tail()
```

```python
accion_hist.dropna(inplace=True)
```

```python
df = accion_hist[["Adj Close"]].copy()
```

```python
df
```

```python
df["Lag_1"] = df["Adj Close"].shift(1)
```

```python
for i in range(1, 29):
  df["Lag_" + str(i)] = df["Adj Close"].shift(i)
```

```python
df["Lag_28"]
```

```python
# Variación porcentual diaria
df["Retorno"] = df["Adj Close"].pct_change()
df
```

```python
df["Sube"] = str(df["Retorno"].copy())
for i in range(len(df["Retorno"])):
  if df["Retorno"][i] > 0:
    df["Sube"][i] = "Sube"
  elif df["Retorno"][i] == 0:
    df["Sube"][i] = "0"
  elif df["Retorno"][i] < 0:
    df["Sube"][i] = "Baja"
```

```python
df
```

```python
df["Retorno_Lag_1"] = df["Lag_1"].pct_change()
df["Retorno_Lag_14"] = df["Lag_14"].pct_change()
df["Retorno_Lag_28"] = df["Lag_28"].pct_change()
```
